```python
def make_pairs(data, groups):
    pairs = []
    labels = []
    classes = len(groups)
    min_n = np.min([len(x) for x in groups])
    
    for c in range(classes):
        for i in range(min_n - 1):
            index_a, index_b = groups[c][i], groups[c][i + 1]
            image_a, image_b = data[index_a], data[index_b]
            pairs += [[image_a, image_b]]
    
            d = np.random.randint(1, classes)
            new_c = (c + d) % classes
            index_a, index_b = groups[c][i], groups[new_c][i]
            image_a, image_b = data[index_a], data[index_b]
            pairs += [[image_a, image_b]]
    
            labels += [1, 0]
    return np.array(pairs), np.array(labels, dtype='float32')


def make_pairs_on_set(data, labels, num_classes):
    groups = [np.where(labels == i)[0] for i in range(num_classes)]
    pairs, new_labels = make_pairs(data, groups)
    return pairs, new_labels


def load_data(dir='./subjects'):
    dir = './subjects'
    data = []
    labels = []
    classes = 0
    for i in os.listdir(dir):
        path = os.path.join(dir, i)
        classes += 1
        for img in os.listdir(path):
            data.append(os.path.join(path, img))
            labels.append(int(i))
    data = np.array(data)
    labels = np.array(labels)
    return make_pairs_on_set(data, labels, classes)
```

```python
class DataGenerator:
    def __init__(self, dir, target_size=None, batch_size=64, normalize=True):
        self.dir = dir
        self.target_size = target_size
        self.batch_size = batch_size
        self.pairs, self.labels = load_data(self.dir)
        self.size = self.labels.shape[0]
        self.normalize = normalize

        # reset index for the first iteration
        self.reset_index()

    def reset_index(self):
        self.index = 0
        self.shuffle_index = np.random.permutation(self.size)
    
    def __iter__(self):
        self.reset_index()
        return self

    def __next__(self):
        if self.index >= self.size:
            raise StopIteration()
            return

        start, end = self.index, min(self.size, self.index + self.batch_size)
        image_pairs = []
        labels = []
        for idx in self.shuffle_index[start:end]:
            # Read image from file
            image_a = cv2.imread(self.pairs[idx][0], cv2.IMREAD_COLOR)
            image_b = cv2.imread(self.pairs[idx][1], cv2.IMREAD_COLOR)
            # convert color from bgr to rgb
            image_a = cv2.cvtColor(image_a, cv2.COLOR_BGR2RGB)
            image_b = cv2.cvtColor(image_b, cv2.COLOR_BGR2RGB)
            # resize the image
            if self.target_size is not None:
                current_size_a = image_a.shape
                current_size_b = image_b.shape
                if current_size_a != self.target_size:
                    image_a = cv2.resize(image_a, self.target_size)
                if current_size_b != self.target_size:
                    image_b = cv2.resize(image_b, self.target_size)
            # append to the image_pairs
            image_pairs += [[image_a, image_b]]
            labels += [self.labels[idx]]

        # Make pairs numpy arrays
        image_pairs = np.array(image_pairs, dtype='float32')
        
        # Normalize return
        if self.normalize:
            mu = np.mean(image_pairs, axis=0)
            std = np.std(image_pairs, axis=0)
            image_pairs = (image_pairs - mu) / std
        # update the index
        self.index = end
        return image_pairs, np.array(labels, dtype='float32')
```

## Building Model

```python
class ConvBlock(Layer):
    def __init__(self, filters, kernel_size=3):
        super(ConvBlock, self).__init__()
        self.conv1 = Conv2D(filters=filters, kernel_size=kernel_size, padding='same', name='conv')
        self.bn = BatchNormalization(axis=-1, name='bn')
        self.act = Activation('relu', name='activation')

    def call(self, input_tensor, training=False):
        x = self.conv1(input_tensor)
        x = self.bn(x, training=training)
        x = self.act(x)
        return x
```

```python
class ResBlock(Layer):
    def __init__(self, filters):
        super(ResBlock, self).__init__()
        self.conv_block1 = ConvBlock(filters[0])
        self.conv_block2 = ConvBlock(filters[1])
        self.conv_block3 = ConvBlock(filters[2])
        self.identity = Conv2D(filters[1], kernel_size=1, padding='same', name='identity')

    def call(self, input_tensor, training=False):
        x = self.conv_block1(input_tensor, training=training)
        x = self.conv_block2(x, training=training)
        skip = self.identity(input_tensor)
        x = self.conv_block3(x + skip, training=training)
        return x
```

```python
class ResNet(Model):
    def __init__(self, output_dim):
        super(ResNet, self).__init__()
        self.conv = Conv2D(filters=64, kernel_size=7, padding='same', name='conv')
        self.bn = BatchNormalization(name='bn')
        self.maxpool = MaxPooling2D(pool_size=(3, 3), name='max_pool')
        self.act = Activation('relu', name='activation')

        self.resblock1 = ResBlock([32, 32, 32])
        self.resblock2 = ResBlock([32, 32, 64])
        self.resblock3 = ResBlock([64, 64, 128])
        self.resblock4 = ResBlock([128, 128, 128])
        self.resblock5 = ResBlock([128, 128, 256])
        self.resblock6 = ResBlock([256, 256, 512])
        self.resblock7 = ResBlock([512, 512, 1024])
        self.globalAvgPool = GlobalAveragePooling2D(name='global_avg_pool')
        self.flatten = Flatten(name='flatten')
        self.dropout = Dropout(0.2, name='dropout')
        self.dense1 = Dense(units=128, name='dense1')
        self.encoded = Dense(units=output_dim, name='encoded')

    def call(self, input_tensor, training=False):
        x = self.conv(input_tensor)
        x = self.bn(x, training=training)
        x = self.act(x)
        x = self.maxpool(x)

        x = self.resblock1(x, training=training)
        x = self.resblock2(x, training=training)
        x = self.resblock3(x, training=training)
        x = self.resblock4(x, training=training)
        x = self.resblock5(x, training=training)
        x = self.resblock6(x, training=training)
        x = self.resblock7(x, training=training)
        x = self.globalAvgPool(x)
        x = self.dropout(x)
        x = self.dense1(x)
        x = self.act(x)
        x = self.dropout(x)
        x = self.encoded(x)
        x = self.act(x)
        return x
```

```python
def euclidean_distance(vects):
    x_1, x_2 = vects
    squared_sum = tf.reduce_sum(tf.square(x_1 - x_2), axis=1, keepdims=True)
    return tf.sqrt(tf.maximum(squared_sum, 1e-8))
```

```python
def euc_dist_output_shape(shapes):
    shape_1, shape_2 = shapes
    return (shape_1[0], 1)
```

```python
def cosine_distance(vects):
    x_1, x_2 = vects
    dots = tf.matmul(x_1, tf.transpose(x_2))
    dots = tf.linalg.diag(dots)
    norm_x1 = tf.linalg.norm(x_1, axis=1)
    norm_x2 = tf.linalg.norm(x_2, axis=1)
    return 1 - tf.divide(dots, (norm_x1 * norm_x2))
```

```python
base_model = ResNet(128)
base_model.build(input_shape=((None, ) + target_size + (3,)))
base_model.summary()
```

```python
class Siamese(Model):
    def __init__(self, base_model):
        super(Siamese, self).__init__()
        self.base_model = base_model
        self.lamb = Lambda(lambda x : euclidean_distance(x), name='output')

    def call(self, inputs, training=False):
        input_a, input_b = inputs[:, 0, :, :, :], inputs[:, 1, :, :, :]
        encoded_a = self.base_model(input_a, training=training)
        encoded_b = self.base_model(input_b, training=training)
        return self.lamb([encoded_a, encoded_b])
```

```python
def contrastive_loss_with_margin(margin):
    def contrastive_loss(y_true, y_pred):
        pred_square = tf.square(y_pred)
        margin_square = tf.square(tf.maximum(margin - y_pred, 0))
        return tf.reduce_mean(pred_square * y_true + (1 - y_true) * margin_square)
    return contrastive_loss
```

```python
optimizer = SGD()
loss_function = contrastive_loss_with_margin(1)
model = Siamese(base_model)
```

```python
@tf.function
def apply_gradients(optimizer, model, pairs, labels):
    with tf.GradientTape() as tape:
        y_pred = model(pairs, training=True)
        loss = loss_function(labels, y_pred)
    grads = tape.gradient(loss, model.trainable_weights)
    grads = zip(grads, model.trainable_weights)
    optimizer.apply_gradients(grads)
    return loss, y_pred
```

```python
def training_step():
    losses = []
    accuracies = []
    total_steps = int(np.ceil(gen.size // gen.batch_size))
    pbar = tqdm(total=total_steps, position=0, leave=True, bar_format='{l_bar}{bar} | {n_fmt}/{total_fmt} | {elapsed}<{remaining}')
    for step, (pairs, labels) in enumerate(gen):
        loss, y_pred = apply_gradients(optimizer, model, pairs, labels)
        losses.append(loss)

        # Calculate the accuracy
        thresholded = np.where(y_pred.numpy().flatten() < 0.5, 1, 0)
        acc = np.sum(thresholded == labels) / thresholded.shape[0]
        
        pbar.set_description(f'Step {step:>4}/{total_steps:>4} | {loss:2.4f} - %{int(acc * 100):>3} ')
        pbar.update()
    return losses, accuracies
```

```python
def validation_step():
    losses = []
    total_steps = int(np.ceil(gen_test.size // gen_test.batch_size))
    pbar = tqdm(total=total_steps, position=0, leave=True, bar_format='{l_bar}{bar} | {n_fmt}/{total_fmt} | {elapsed}<{remaining}')
    for step, (pairs, labels) in enumerate(gen_test):
        logits = model(pairs, training=False)
        loss = loss_function(labels, logits)
        losses.append(loss)
        pbar.set_description(f'Step {step:>4}/{total_steps:>4} | {loss:2.4f} ')
        pbar.update()
    return losses
```

```python
epochs = 2
avg_loss = []
avg_test_losses = []

for epoch in range(epochs):
    print(f'Start of epoch {epoch + 1: >2}')
    losses = training_step()
    test_losses = validation_step()
    avg_loss.append(np.mean(losses))
    avg_test_losses.append(np.mean(test_losses))
    print(f'epoch: {epoch + 1:>2}, loss: {avg_loss[-1]:1.3e}, test_loss: {avg_test_losses[-1]:1.3e}')
```
